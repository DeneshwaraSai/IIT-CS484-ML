```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import math
```

```python
data = pd.read_csv('lab03_dataset_2.csv')
data.head()
```

#### 1) The input features used during training should be subjected to feature scaling, specifically the min-max scaling. (2 marks)

```python
minMaxScaler = MinMaxScaler()
scaledData = minMaxScaler.fit_transform(data.iloc[:, 2:])
data.iloc[:, 2:]
```

```python
scaledInfo = pd.DataFrame(scaledData, columns=data.columns[2:])
scaledInfo.head()
```

```python
X = scaledInfo
Y = data[['FRAUD']]
```

```python
Y.head()
```

#### Use the first 20% of the dataset i.e., the first 20% of the rows as the test set, while the remaining bottom 80% rows will be your training set.

```python
print(list(data.shape))

splitNum = math.ceil(0.2 * list(data.shape)[0])

X_train = scaledInfo.iloc[splitNum:, :] # Use it for train purpose
X_test = scaledInfo.iloc[:splitNum, :] # Use it for test purpose

Y_train = Y.iloc[splitNum:, :]
Y_test = Y.iloc[:splitNum, :] 
```

```python
X_test
```

#### 2) You will use sklearnâ€™s k â€“ nearest neighbors module to learn a classification model with multiple nearest neighbors ranging from 2 to 5. Apply the learned kâ€“NN model to classify the test set. Compute the misclassification rates for k ranging from 2 to 5. Use Euclidean distance as the similarity measure. (5 marks)

```python
for i in range(2, 6):
    knn = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
    knn.fit(X_train.squeeze(), Y_train.squeeze())
    y_pred = knn.predict(X_test)
    print("------------------------------------------------------------------------------------------------------------------------------------")
    print("The n-neighbour is :", i)
    print(y_pred,  len(y_pred))
    misclassification = 1 - accuracy_score(Y_test, y_pred)
    print(f"The misclassification when k = {i}, is {misclassification:.4f}")
    print("------------------------------------------------------------------------------------------------------------------------------------\n")
```

#### 3) Next, apply sklearnâ€™s kâ€“d tree module to classify the test set. In a similar manner to the above scenario, compute the misclassification rates for k ranging from 2 to 5. Use Manhattan distance as the similarity measure. (5 marks)

```python
for i in range(2, 6):
    print("-----------------------------------------------------------------------------------------------------------------------------------")
    knn = KNeighborsClassifier(n_neighbors=i, algorithm='kd_tree', metric='manhattan')
    knn.fit(X_train.squeeze(), Y_train.squeeze())
    y_pred = knn.predict(X_test)
    print("n_neighbors :",i)
    accuracyScore = 1 - accuracy_score(Y_test, y_pred)
    print(f"The n_neighbors are : {i} and its misclassification is : {accuracyScore:.4f}")
    print("-----------------------------------------------------------------------------------------------------------------------------------\n")
```
